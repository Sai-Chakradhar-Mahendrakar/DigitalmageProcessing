```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to display images using matplotlib
def display_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# a. Read, Display, and Write a Color Image in Other Formats
image = cv2.imread('Images/earth.jpg')  # Read the image
display_image(image, 'Original Image')  # Display the image using matplotlib

cv2.imwrite('output_image.png', image)  # Save the image in another format (e.g., PNG)
```

```python
# b. Find RED, GREEN, and BLUE Planes of the Color Image
blue_plane = image[:, :, 0]  # Blue channel
green_plane = image[:, :, 1]  # Green channel
red_plane = image[:, :, 2]  # Red channel

display_image(cv2.merge([blue_plane, blue_plane, blue_plane]), 'Blue Plane')
display_image(cv2.merge([green_plane, green_plane, green_plane]), 'Green Plane')
display_image(cv2.merge([red_plane, red_plane, red_plane]), 'Red Plane')
```

```python
# c. Convert Color Image to Grayscale Image and Binary Image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)  # Convert to binary

plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```

```python
# d. Resize the Image by One Half and One Quarter
image_half = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)  # Half size
image_quarter = cv2.resize(image, (0, 0), fx=0.25, fy=0.25)  # Quarter size

display_image(image_half, 'Half Size Image')
display_image(image_quarter, 'Quarter Size Image')
```

```python
# e. Rotate the Image by 45, 90, and 180 Degrees
def rotate_image(img, angle):
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(img, M, (w, h))
    return rotated
```

```python
image_45 = rotate_image(image, 45)
image_90 = rotate_image(image, 90)
image_180 = rotate_image(image, 180)

display_image(image_45, '45 Degree Rotation')
display_image(image_90, '90 Degree Rotation')
display_image(image_180, '180 Degree Rotation')
```
